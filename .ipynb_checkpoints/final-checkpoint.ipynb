# a set of unique words
I took The Waves by Virginia Woolf, and cross-compared that with a set of the most common English words from the corpora Github repo--and created a new list of less common English words from The Waves.

```python
book = open("./thewaves.txt").read()
text = book.replace("\n", " ").split(" ")
text = [t.strip(string.punctuation) for t in text]
t1 = set(text)
t2 = None
with open('./common.json') as commonWords:
    t2 = json.load(commonWords)["commonWords"]
    t2.extend(['', 'her', 'him', 'my', 'hers', 'his', 'mine', 'I', 'your', 'yours', 'they', 'their', 'theirs']) # extra special words to omit from final words list
words = [x for x in t1 if x not in t2]
print(words)
print(len(words))
```

## words that appeared at least twice
I ran into an issue where a lot of the double entendres I found in the text were only used once. Since I wanted to make the poem out of two different usages of the word, it would require the word to have appeared more than once in the text, so I narrowed the list further.

```python
count = Counter(book.split(" "))
words = [w for w in words if count[w] > 1]
print(len(words))
```

# finding double entendres
using the sample word "china" which is closest to what I mean by a double entendre--it has multiple entries in the Noun category<br><br>
I'll use this as a rule for determining which words in my word list are double entendres--if they have more than 2 (to be safe) entries in the Noun category then I'll count it as a double entendre

```python
dictionary=PyDictionary("china")

print(dictionary.printMeanings()) 
print(dictionary.getMeanings())
```

# getting the double entendres
the dictionary API takes a lot longer than I expected, so I'll shorten the call by only calling the API until I get at least 5 double entendres

```python
doubleEntendres = []
random.shuffle(words)
wordIndex = 0
while len(doubleEntendres) < 20 and wordIndex < len(words):
    word = words[wordIndex]
    dictionary=PyDictionary(word)
    meanings = dictionary.getMeanings()
    if meanings and meanings[word] and "Noun" in meanings[word] and len(meanings[word]["Noun"]) > 3:
        doubleEntendres.append(word)
        print(word + ' is a double entendre')
#         print(meanings)
    else:
        print(word + ' is NOT a double entendre')
    wordIndex += 1
print(doubleEntendres)
# dictionary=PyDictionary(words[1])
# dictionary.getMeanings()
# dictionary.printMeanings()
```

# taking the words and placing it back into context
then I took the words and found lines with those words in Woolf's text, and created 'mirror' images out of them

```python
# entendresList = ['bolt','frame','barks','cutting','fibre','gutter','hoop','earth']
entendresList=doubleEntendres
printReverse = False
tabs =  [10,9,8,7,6,6,6,6,6,6,6,6,6,6,6,6,6,6,6,6,6,6,6,6,6,6,6,6]
tabs2 = [14,14,11,8,5,5,5,5,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4]
fib =   [0, 1, 5, 8, 13, 13, 13, 13, 13,13,13,13,13,13,13,13,13,13,13,13,13,13,13,13,13,13,13,13,13]
mir_dict = {"a":"ɒ","b": "d","c": "ɔ","d": "b","e": "ɘ","f": "ʇ","g": "ǫ","h": "ʜ","i": "i","j": "Ⴑ","k": "ʞ","l": "l","m": "m","n": "n","o": "o","p": "q","q": "p","r": "ɿ","s": "ƨ","t": "ƚ","u": "u","v": "v","w": "w","x": "x","y": "y","z": "z"}
```

*note: you can edit the printReverse boolean above to have the code below print the 'mirrored' text<br><br>
the code below might no be very readable but having almost lost all my code--I learned that you can retrieve your last python commands in jupyter notebook in the form of a sqlite file however, it only saves up to about the first ~20 lines or so of code in each cell, so I'm condensing my code a bit just in case

```python
for entend in entendresList:
    print('\033[1m' + entend +  '\033[0m')
    eindex = book.find(entend)
    left = book[eindex-150:eindex+500].replace('\n',' ')
    nindex = book.find(entend,eindex+200)
    right = book[nindex-150:nindex+500].replace('\n',' ')
    leftIndex = 0
    rightIndex = 0
    for i,t in enumerate(tabs):
        inc = tabs[i]+fib[i]          # LEFT SIDE
        print('_' * t, end='')
        if (i == 9):
            print('\033[1m' + left[leftIndex:leftIndex+inc] +  '\033[0m', end='')
        else:
            print(left[leftIndex:leftIndex+inc], end='')
        leftIndex += inc
        print('_' * tabs2[i], end='') # RIGHT SIDE
        if printReverse:
            res = [mir_dict[ele] if ele in mir_dict else ele for ele in right[rightIndex:rightIndex+inc][::-1]]
            print(''.join(res), end="\n")
        else:
            if (i == 9):
                print('\033[1m' + right[rightIndex:rightIndex+inc] +  '\033[0m')
            else:
                print(right[rightIndex:rightIndex+inc])
        rightIndex += inc
    print("\n")
    print("******************************************************", end="\n\n\n")
```
